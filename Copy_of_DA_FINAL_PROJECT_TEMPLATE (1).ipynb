### Converting PDF Dashboard to Image for Display

Since interactive PDF dashboards cannot be directly embedded in Colab, one common approach is to convert the PDF into an image. This allows you to display a static representation of the dashboard's pages within your notebook.

We will use the `PyMuPDF` library (also known as `fitz`) for this conversion. First, let's install it.

Now, let's convert the first page of your `Main project dashboard.pdf` into an image and display it. If your dashboard has multiple pages, you can iterate through them to display each one.

```python
import fitz # PyMuPDF
from PIL import Image
import io

pdf_path = '/content/Main project dashboard (1).pdf'

try:
    # Open the PDF document
    doc = fitz.open(pdf_path)

    # Get the first page (page 0)
    page = doc.load_page(0)

    # Render the page as a high-resolution PNG image
    # zoom_x and zoom_y factors increase the resolution
    zoom_x = 2.0  # horizontal zoom
    zoom_y = 2.0  # vertical zoom
    mat = fitz.Matrix(zoom_x, zoom_y)
    pix = page.get_pixmap(matrix=mat)

    # Convert to a PIL Image
    img = Image.open(io.BytesIO(pix.tobytes("png")))

    # Display the image
    display(img)

    doc.close()

except FileNotFoundError:
    print(f"Error: The file '{pdf_path}' was not found. Please ensure it's uploaded or correctly specified.")
except Exception as e:
    print(f"An error occurred: {e}")
```
